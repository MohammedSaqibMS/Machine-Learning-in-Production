# Deploying a deep learning model with Docker and a cloud service

In these optional labs, you will deploy the application you built via Docker and AWS Elastic Beanstalk. Your model will now move out of the Jupyter notebook and be exposed to external traffic just like you would do in real-world applications!

**Note that this exercise is optional; you don't have to complete this item to complete this course.**

Click on `docker-deploy.ipynb` on the left file explorer to get started.
